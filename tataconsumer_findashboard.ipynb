```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("C:/Users/HP/Downloads/tata_consumer_financials.csv")
```

```python
print("Dataset Preview:")
print(df.head())
```

```python
df['Date'] = pd.to_datetime(df['Year'].astype(str) + '-' + df['Month'])
df.sort_values(by='Date', inplace=True)

# Aggregate Monthly Revenue
monthly_revenue = df.groupby(['Year', 'Month'])['Revenue'].sum().reset_index()

# Aggregate Monthly EBITDA
monthly_ebitda = df.groupby(['Year', 'Month'])['EBITDA'].sum().reset_index()
```

```python
# Year on Year Growth
revenue_yearly = df.groupby('Year')['Revenue'].sum()
yoy_growth = revenue_yearly.pct_change() * 100

print("Yearly Revenue:")
print(revenue_yearly)
print("\nYear-over-Year Growth (%):")
print(yoy_growth)
```

```python
# EBITDA Margin %
df["EBITDA_Margin_%"] = (df["EBITDA"] / df["Revenue"]) * 100
```

```python
# CAGR Calculation
years = len(df) - 1
revenue_cagr = ((df["Revenue"].iloc[-1] / df["Revenue"].iloc[0]) ** (1/years) - 1) * 100
```

```python
#Financial Ratios

# Net Profit Margin
df["Net_Profit_Margin_%"] = (df["Net_Profit"] / df["Revenue"]) * 100

# Debt to Equity Ratio
df["Debt_to_Equity"] = df["Total_Liabilities"] / df["Equity"]

# Return on Equity (ROE)
df["ROE_%"] = (df["Net_Profit"] / df["Equity"]) * 100
```

```python
print("Revenue CAGR (2020-2024): {:.2f}%".format(revenue_cagr))
print("\nLatest Year Financial Ratios:")
print(df.tail(1)[[
    "EBITDA_Margin_%", 
    "Net_Profit_Margin_%", 
    "Debt_to_Equity", 
    "ROE_%"
]])
```

```python
# Revenue & EBITDA Trend
plt.figure()
plt.plot(df["Year"], df["Revenue"])
plt.plot(df["Year"], df["EBITDA"])
plt.title("Revenue & EBITDA Trend")
plt.xlabel("Year")
plt.ylabel("Amount")
plt.show()
```

```python
# EBITDA Margin Trend
plt.figure()
plt.plot(df["Year"], df["EBITDA_Margin_%"])
plt.title("EBITDA Margin % Trend")
plt.xlabel("Year")
plt.ylabel("EBITDA Margin (%)")
plt.show()
```

```python
# ROE Trend
plt.figure()
plt.plot(df["Year"], df["ROE_%"])
plt.title("Return on Equity (%)")
plt.xlabel("Year")
plt.ylabel("ROE (%)")
plt.show()
```
